```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Rolando.serrano\Documents\informes\retornos\nexoya14092021.csv',encoding='latin-1',sep=';')
```

```python
df['FECHA RECEPCION'].sample(3)
```

```python
df['FECHA RECEPCION'] = pd.to_datetime(df['FECHA RECEPCION'], format = '%d-%m-%y')
df['month'] = pd.DatetimeIndex(df['FECHA RECEPCION']).month
df['year'] = pd.DatetimeIndex(df['FECHA RECEPCION']).year
```

```python
per = df['FECHA RECEPCION'].dt.to_period('M')
g = df.groupby(per)
g = g.sum().reset_index()
g[['FECHA RECEPCION','MONTO RETORNO']]
```

```python
g = g.rename({'FECHA RECEPCION': 'RECEPCION'},axis=1)
g['RECEPCION'] = g['RECEPCION'].dt.strftime('%Y-%m')
```

```python
d = g[26:]
d = d.drop([47,48])
G =px.bar(d, x= 'RECEPCION', y='MONTO RETORNO',color='RECEPCION')
G.show()
```

```python
cero = df[df['MONTO RETORNO'] == 0]

per = cero['FECHA RECEPCION'].dt.to_period('M')
g = df.groupby(per)
g = g.sum().reset_index()
g[['FECHA RECEPCION','MONTO RETORNO', 'TOTAL RECUPERABLE']]
g = g.rename({'FECHA RECEPCION': 'RECEPCION'},axis=1)
g['RECEPCION'] = g['RECEPCION'].dt.strftime('%Y-%m')
```

```python
g = g[18:]
G =px.bar(g, x= 'RECEPCION', y='TOTAL RECUPERABLE')
G.show()
```
